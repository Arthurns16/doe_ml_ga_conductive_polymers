```
import sys
import numpy as np
import pickle
import matplotlib.pyplot as plt
import os

# Fix for numpy._core mapping
sys.modules['numpy._core'] = np.core

# Load metrics
with open('history_metrics.pkl', 'rb') as f:
    metrics = pickle.load(f)

# Extract and slice first 20 generations
best = metrics['best_history'][:20]
avg = metrics['avg_history'][:20]
mut_cat = metrics['mut_cat_hist'][:20]
mut_cont = metrics['mut_cont_hist'][:20]
diversity = metrics['diversity_hist'][:20]

# Plot graphs with English labels
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Fitness over first 20 generations
axes[0].plot(best, label="Best", linewidth=2)
axes[0].plot(avg,  label="Average", linestyle='--')
axes[0].set_title("Fitness over Generations")
axes[0].set_xlabel("Generation")
axes[0].set_ylabel("Fitness")
axes[0].legend()

# Mutation rates over first 20 generations
axes[1].plot(mut_cat, label="Categorical Mutation Rate", linewidth=2)
axes[1].plot(mut_cont, label="Continuous Mutation Rate", linestyle='--')
axes[1].set_title("Mutation Rates")
axes[1].set_xlabel("Generation")
axes[1].set_ylabel("Rate")
axes[1].legend()

# Diversity over first 20 generations
axes[2].plot(diversity, label="Normalized Shannon Index", linewidth=2)
axes[2].set_title("Population Diversity")
axes[2].set_xlabel("Generation")
axes[2].set_ylabel("Diversity")
axes[2].legend()

# Adjust layout and save once before show
fig.tight_layout()
output_path = os.path.join(os.getcwd(), "history_metrics_first20.png")
fig.savefig(output_path, dpi=300, bbox_inches="tight")
print(f"Figure saved to: {output_path}")

plt.show()

```
